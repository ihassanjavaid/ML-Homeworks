```python
# Conventional Way Regression

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 0, 2, 6, 4, 3, 3])
y = np.array([95, 90, 90, 55, 70, 80, 85])

x_sq = np.square(x)

xy_sumission = np.sum(np.dot(x, y))

n = len(x)

m = ( (n*xy_sumission) - np.dot(np.sum(x),np.sum(y)) ) / ( (n*np.sum(x_sq)) - np.square(np.sum(x)) )

b = ( np.sum(y)/n ) - (m*(np.sum(x)/n))

y_ans = m*x + b

print('y = mx + b')

print(f'm = {m}')
print(f'b = {b}')

print(f'First Value: {y_ans[0]}')

plt.scatter(x, y)
plt.plot(x, y_ans)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Sci-kit Regression

from sklearn.linear_model import LinearRegression

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

lr = LinearRegression()

lr.fit(x, y)

# value of m
print(f'm: {lr.coef_[0][0]}')

# value of b
print(f'b: {lr.intercept_[0]}')

# value of y[0]
print(f'First Value: {lr.predict(x[[0]])[0][0]}')

plt.scatter(x, y)
plt.plot(x, lr.predict(x))
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
