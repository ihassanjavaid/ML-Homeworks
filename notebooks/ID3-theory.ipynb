```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

names = ['Restaurants','Meal','Day','Cost','Reaction']
dataset = pd.read_csv('rest.csv', names=names)

dataset.head()
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
id3 = DecisionTreeClassifier(criterion='entropy')
id3.fit(x, y)

from sklearn import tree

tree.plot_tree(id3)

tree.export_graphviz(id3,out_file='tree.dot')
```
